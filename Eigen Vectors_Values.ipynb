```python
m_trim = img[1:690, 1:690]
```

```python
from skimage import io, color
```

```python
#lina_color = plt.imread(img)
img_2d = color.rgb2gray(m_trim)
print(img_2d)
```

```python
plt.imshow(img_2d)
plt.colorbar()
np.shape(img_2d)
```

```python
import scipy.linalg as la
val,vec=la.eig(img_2d)
```

```python
real_vec=vec.real
imag_vec=vec.imag

real_val=val.real
imag_val=val.imag
```

```python
fig,ax=plt.subplots(2,figsize=(5,5))
ax[0].imshow(real_vec)
ax[1].imshow(imag_vec)
plt.text(400,100,'Real',color='white')
plt.tight_layout()
```

```python
fig,ax=plt.subplots(4,figsize=(5,5))
ax[0].plot(real_vec[500,:])
ax[1].plot(imag_vec[500,:])
ax[2].plot(real_val)
ax[3].plot(imag_val)
plt.tight_layout()
```
