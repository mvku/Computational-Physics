# How to show the series of images in a single window ?

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
w=10
h=10
fig=plt.figure(figsize=(8, 8))
gs1 = gridspec.GridSpec(8, 8)
gs1.update(wspace=0.025, hspace=0.05)
columns = 4
rows = 4
for i in range(1, columns*rows +1):
    img = np.random.randint(10, size=(h,w))
    fig.add_subplot(rows, columns, i)
    #plt.text(0.5,0.8,'Image'+'%02d'%i,bbox=dict(facecolor='yellow', alpha=1))
    plt.axis('on')
    plt.imshow(img)
    plt.tight_layout()
#plt.subplots_adjust(wspace=0.05, hspace=0.05)
plt.show()
```

# Remove the axis

```python
fig=plt.figure(figsize=(8, 8))
gs1 = gridspec.GridSpec(8, 8)
gs1.update(wspace=0.025, hspace=0.05)
columns = 4
rows = 4
for i in range(1, columns*rows +1):
    img = np.random.randint(10, size=(h,w))
    fig.add_subplot(rows, columns, i)
    #plt.text(0.5,0.8,'Image'+'%02d'%i,bbox=dict(facecolor='yellow', alpha=1))
    plt.axis('off')# remove the axis
    plt.imshow(img)
    plt.tight_layout()
#plt.subplots_adjust(wspace=0.05, hspace=0.05)
plt.show()
```

# Remove spaces between the images

```python
fig=plt.figure(figsize=(8, 8))
gs1 = gridspec.GridSpec(8, 8)
gs1.update(wspace=0.025, hspace=0.05)
columns = 4
rows = 4
for i in range(1, columns*rows +1):
    img = np.random.randint(10, size=(h,w))
    fig.add_subplot(rows, columns, i)
    #plt.text(0.5,0.8,'Image'+'%02d'%i,bbox=dict(facecolor='yellow', alpha=1))
    plt.axis('off')# remove the axis
    plt.imshow(img)
    plt.tight_layout()
plt.subplots_adjust(wspace=0.01, hspace=0.01)
plt.show()
```

# Adding text in the images

```python
fig=plt.figure(figsize=(8, 8))
gs1 = gridspec.GridSpec(8, 8)
gs1.update(wspace=0.025, hspace=0.05)
columns = 4
rows = 4
for i in range(1, columns*rows +1):
    img = np.random.randint(10, size=(h,w))
    fig.add_subplot(rows, columns, i)
    plt.text(0.5,0.8,'Image'+'%02d'%i,bbox=dict(facecolor='yellow', alpha=1))
    plt.axis('off')# remove the axis
    plt.imshow(img)
    plt.tight_layout()
plt.subplots_adjust(wspace=0.01, hspace=0.01)
plt.show()
```

# Change the text background color

```python
fig=plt.figure(figsize=(8, 8))
gs1 = gridspec.GridSpec(8, 8)
gs1.update(wspace=0.025, hspace=0.05)
columns = 4
rows = 4
for i in range(1, columns*rows +1):
    img = np.random.randint(10, size=(h,w))
    fig.add_subplot(rows, columns, i)
    plt.text(0.5,0.8,'Image'+'%02d'%i,bbox=dict(facecolor='white', alpha=1))
    plt.axis('off')# remove the axis
    plt.imshow(img)
    plt.tight_layout()
    
plt.subplots_adjust(wspace=0.01, hspace=0.01)
plt.show()
```

# Changing the image background color variation

```python
fig=plt.figure(figsize=(8, 8))
gs1 = gridspec.GridSpec(8, 8)
gs1.update(wspace=0.025, hspace=0.05)
columns = 4
rows = 4
for i in range(1, columns*rows +1):
    img = np.random.randint(5, size=(h,w))
    fig.add_subplot(rows, columns, i)
    plt.text(0.2,0.8,'Image '+'%02d'%i+' Number', bbox=dict(facecolor='white', alpha=1))
    plt.axis('off')# remove the axis
    plt.imshow(img)
    plt.tight_layout()
plt.subplots_adjust(wspace=0.01, hspace=0.01)
plt.show()
```
