```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series  # for convenience

import pims
import trackpy as tp
```

```python
@pims.pipeline
def gray(image):
    return image[:, :, 1]  # Take just the green channel
sub=0
frames = gray(pims.open('C:/Users/vijay/Downloads/ants/*.jpg'))
for image in frames:
    a=np.sum(image)
```

```python
micron_per_pixel = 0.15192872980868
feature_diameter = 2.12 # um
radius = int(np.round(feature_diameter/2.0/micron_per_pixel))
if radius % 2 == 0:
    radius += 1
print('Using a radius of {:d} px'.format(radius))
a-frames[0]
```

```python
f_locate = tp.locate(frames[0]-a, radius+2, minmass=200)
```

```python
tp.annotate(f_locate, frames[0]-a, plot_style={'markersize': radius});
```
